```python
# Clean the worldwide_gross column
gross_budgets['worldwide_gross'] = gross_budgets['worldwide_gross'].replace('$', '').replace(',', '')

# Convert the column to float
gross_budgets['worldwide_gross'] = gross_budgets['worldwide_gross'].astype(float)

# Clean the production_budget column
gross_budgets['production_budget'] = gross_budgets['production_budget'].replace('$', '').replace(',', '')

# Convert the column to float
gross_budgets['production_budget'] = gross_budgets['production_budget'].astype(float)
```

```python
# Calculate the profit column
gross_budgets['profit'] = gross_budgets['worldwide_gross'] - gross_budgets['production_budget']
# View the updated dataset
print(gross_budgets[['movie','profit']])
```

```python
# Merge the two dataframes
movies_w_budgets = pd.merge(movies_df, movie_budgets_df, on=['movie'])
movies_w_budgets
```

```python
movies_w_budgets = movies_w_budgets.drop_duplicates(subset=['movie'])
movies_w_budgets
```

```python
# Assuming the money value is stored as a string in the column 'worldwide_gross'
movies_w_budgets['worldwide_gross'] = movies_w_budgets['worldwide_gross'].str.replace('$', '').str.replace(',', '')

# Convert the column to integer
movies_w_budgets['worldwide_gross'] = movies_w_budgets['worldwide_gross'].astype(float)

# Print the cleaned and converted column
print(movies_w_budgets['worldwide_gross'])
```

```python
import matplotlib.pyplot as plt

# Sort the DataFrame by worldwide_gross in descending order
sorted_df = movies_w_budgets.sort_values('worldwide_gross', ascending=False)

# Select the top 10 movies
top_10_movies = sorted_df.head(10)

# Plot the bar chart
plt.figure(figsize=(12, 6))
plt.bar(top_10_movies['movie'], top_10_movies['worldwide_gross'])
plt.xticks(rotation=45)
plt.xlabel('Movie')
plt.ylabel('Worldwide Gross in Billions')
plt.title('Top 10 Grossing Movies')
plt.show()
```

```python
import zipfile
import sqlite3
import pandas as pd

# Path to the zip file
zip_file_path = '../../data/im.db.zip'
# Folder to extract the zip file contents
extract_folder = 'C:/Users/rmasters/Documents/Flatiron/Unzipped_database'

# Open the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Extract the contents of the zip file to the extract_folder
    zip_ref.extractall(extract_folder)

# Connect to the database file
db_file_path = f'{extract_folder}/im.db'
conn = sqlite3.connect(db_file_path)

# Create a cursor object to execute SQL queries
cursor = conn.cursor()

# Query the database and fetch the result into a pandas DataFrame
movie_ratings_table = """
SELECT *
FROM movie_ratings
;"""

movie_ratings_df = pd.read_sql_query(movie_ratings_table, conn)
movie_ratings_df
```
